## **Control flow**
 * if
 * for
 * while


````python
# if
score = 82
if score >= 80:
    print("passed")
else:
    print("failed")
````

````python
# multiple grades
score = 80

if score >= 80:
    print("A")
elif score > 70:
    print("B")
elif score > 60:
    print("C")
else:
    print("D")
````

````python
# gets input from users
username = input("whats 's your name: ")
age = int(input("how old are you: "))
````

````python
# multiple conditions
# night club
has_id = True
age_over_18 = False
vip_customer = False

if has_id and (age_over_18 or vip_customer) :
    print("Welcome to the club")
else:
    print("Please come next time.")
````

````python
# for & while
items = ["egg","milk","bread"]   #plural

for item in items:    # singular in plural format
    if item == "milk":
        print("CP Meiji")
    else:
        print(item)
````

````python
## list comprehension
result = []

for item in items:
    result.append(item.upper())
print(result)
````

````python
items
````

````python
#list comprehension
[item.upper() for item in items]
````

````python
## while
while True:
    user_input = input("Play on Y/N: ")
    if user_input == "N":
        print("Goodbye!")
        break
    else:print("Goooo!")
````

````python
## while another example
count = 0
while count < 5:
    print("hi!")
    count += 1
````

## **OOP**

Object Oriented Programming

````python
class Employee():
    """
    A new employee at Google
    """

    def __init__(self, name, age, city):
        self.name = name
        self.age = age
        self.city = city

    def working(self):
        print("I'm working now!")
````
